```python
from sklearn.svm import LinearSVC
Previsao_valores_LinearSVC = list()
Data_Previsao_LinearSVC = list()

#Aplicando OrdinalEncoder
enc = OrdinalEncoder()

Base[["Codigo Fipe","Marca", "Modelo", "Ano do Modelo", "Ms Referencia", "Ano Referencia"]] = enc.fit_transform(Base[["Codigo Fipe","Marca", "Modelo", "Ano do Modelo", "Ms Referencia", "Ano Referencia"]])
X = Base[Base.columns[0:6]].values
y = Base.loc[:,'Valor'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
X_train.shape
X_test.shape
X_train.shape[0] + X_test.shape[0]

modelo = LinearSVC()
modelo.fit(X_train, y_train)

print('R = {} Quanto a linha de regresso ajusta-se aos dados'.format(modelo.score(X_train, y_train).round(4)))
### Gerando previses para os dados de TESTE (X_test) utilizando o mtodo *predict()* do objeto "modelo"
y_previsto = modelo.predict(X_test)
### Obtendo o coeficiente de determinao (R) para as previses do nosso modelo
print('R = %s Previses do nosso modelo' % metrics.r2_score(y_test, y_previsto).round(4))


##########################################################
### Criando um simulador simples
Codigo_Fipe=0.0
Marca=0.0
Modelo=0.0
Ano_do_Modelo=0.0
#simulando tabela fipe do ms de setembro
Ms_Referencia=8.0
#simulando tabela fipe do ano de 2022
Ano_Referencia=15.0
entrada=[[Codigo_Fipe, Marca, Modelo, Ano_do_Modelo, Ms_Referencia, Ano_Referencia ]]
Set = modelo.predict(entrada)[0]

Data_Previsao_LinearSVC.append('01/09/2022')
Set = Set.round(2)
Previsao_valores_LinearSVC.append(Set)




##########################################################
### Criando um simulador simples
Codigo_Fipe=0.0
Marca=0.0
Modelo=0.0
Ano_do_Modelo=0.0
#simulando tabela fipe do ms de outubro
Ms_Referencia=9.0
#simulando tabela fipe do ano de 2022
Ano_Referencia=15.0
entrada=[[Codigo_Fipe, Marca, Modelo, Ano_do_Modelo, Ms_Referencia, Ano_Referencia ]]
Out = modelo.predict(entrada)[0]

Data_Previsao_LinearSVC.append('01/10/2022')
Out = Out.round(2)
Previsao_valores_LinearSVC.append(Out)



##########################################################
### Criando um simulador simples
Codigo_Fipe=0.0
Marca=0.0
Modelo=0.0
Ano_do_Modelo=0.0
#simulando tabela fipe do ms de novembro
Ms_Referencia=10.0
#simulando tabela fipe do ano de 2022
Ano_Referencia=15.0
entrada=[[Codigo_Fipe, Marca, Modelo, Ano_do_Modelo, Ms_Referencia, Ano_Referencia ]]
Nov = modelo.predict(entrada)[0]

Data_Previsao_LinearSVC.append('01/11/2022')
Nov = Nov.round(2)
Previsao_valores_LinearSVC.append(Nov)

##########################################################



Codigo_Fipe=0.0
Marca=0.0
Modelo=0.0
Ano_do_Modelo=0.0
#simulando tabela fipe do ms de dezembro
Ms_Referencia=11.0
#simulando tabela fipe do ano de 2022
Ano_Referencia=15.0
entrada=[[Codigo_Fipe, Marca, Modelo, Ano_do_Modelo, Ms_Referencia, Ano_Referencia ]]
Dez = modelo.predict(entrada)[0]

Data_Previsao_LinearSVC.append('01/12/2022')
Dez = Dez.round(2)
Previsao_valores_LinearSVC.append(Dez)

#unificando listas
Previsao_Df_LinearSVC = list(zip(Data_Previsao_LinearSVC,Previsao_valores_LinearSVC))

#criando dataframe pandas
estrutura_df = {
    'Data': str,
    'Valores': str
}

Previsao_Df_LinearSVC = pd.DataFrame(Previsao_Df_LinearSVC, dtype=str)
Previsao_Df_LinearSVC.columns = estrutura_df.keys()
Previsao_Df_LinearSVC = Previsao_Df_LinearSVC.astype(estrutura_df)

#transformando coluna em data
Previsao_Df_LinearSVC["Data"] = pd.to_datetime(Previsao_Df_LinearSVC["Data"])
#sequencia dia mes ano
Previsao_Df_LinearSVC['Data'] = Previsao_Df_LinearSVC['Data'].dt.strftime('%m-%d-%Y')
import seaborn as sns 
sns.scatterplot(x="Data", y="Valores", data=Previsao_Df_LinearSVC)
Previsao_Df_LinearSVC

```
